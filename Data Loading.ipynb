### 8. Concatenate Data

```python
data1 = data[['Survived', 'Pclass']]
data1
```

```python
data2 = data[['Sex', 'Age']]
data2
```

#### concat

```python
data_concat = [data1, data2]
data3 = pd.concat(data_concat, axis = 1)
```

```python
data3
```

```python
# Apabila pemilihan aksis salah, bisa menyebabkan kesalahan pada data
data4 = pd.concat(data_concat, axis = 0)
data4
```

```python
# Apabila axis tidak diisi (default: Axis = 0)
data5 = pd.concat(data_concat)
data5
```
