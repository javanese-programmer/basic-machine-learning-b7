# Data Visualization

### 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Load Data

```python
data = pd.read_csv("iris.csv")
```

```python
data.head()
```

```python
# Ubah nama kolom DataFrame
columns = {'SepalLengthCm' : 'Sepal_Length(cm)', 'SepalWidthCm' : 'Sepal_Width(cm)', 'PetalLengthCm' : 'Petal_Length(cm)', 'PetalWidthCm':'Petal_Width(cm)'}
```

#### rename

```python
data = data.rename(columns=columns)
```

```python
data.head()
```

#### drop

```python
# Membuang kolom yang tidak dipakai
data = data.drop(['Id'], axis = 1)
```

```python
data.head()
```

#### unique

```python
# Mencari nilai unik dari suatu kolom
data['Sepal_Length(cm)'].unique()
```

```python
data['Species'].unique()
```

#### dtypes

```python
data.dtypes
```

#### info

```python
data.info()
```

### 3. Data Visualization

#### histogram

```python
data['Species'].hist()
```

```python
data['Species'].hist()
plt.xlabel('Species')
plt.show()
```

#### Scatter Plot

```python
data.plot(kind = 'scatter', x = 'Sepal_Length(cm)', y = 'Sepal_Width(cm)')
plt.show()
```

```python
help(plt.plot)
```

```python
data.plot(kind = 'scatter', x = 'Petal_Length(cm)', y = 'Petal_Width(cm)')
plt.show()
```

#### Distribution Plot

```python
sns.displot(data['Sepal_Length(cm)'], bins=10)
plt.show()
```

```python
# bins untuk menentukan jumlah kelompok dalam distribusi
sns.displot(data['Sepal_Length(cm)'], bins = 20)
plt.show()
```

```python
sns.displot(data['Petal_Length(cm)'], bins = 20)
plt.show()
```

#### Box Plot

```python
# Dengan matplotlib
data.plot(kind = 'box')
plt.xlabel('features')
plt.ylabel('Length(cm)')
plt.show()
```

```python
# Dengan Seaborn
sns.boxplot(data = data)
plt.xlabel('features')
plt.ylabel('Length(cm)')
plt.show()
```

```python
# Menspesifikkan kolom yang ditampilkan pada sumbu x dan y
sns.boxplot(data=data, x='Species', y ='Sepal_Width(cm)')
plt.xlabel('Species')
plt.ylabel('Sepal Width in cm')
plt.show()
```

#### Multivariable : Pair Plot

```python
# hue untuk menspesifikkan kategori yang diberi indikator warna
sns.pairplot(data = data, hue = 'Species')
plt.show()
```

#### Multivariable : Heat Map (Show Correlation)

```python
sns.heatmap(data=data.corr(), annot=True)
plt.show()
```
