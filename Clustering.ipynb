# Clustering

## 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# import dataset
from sklearn.datasets import make_blobs
```

## 2. Generate Datasets

```python
X, _ = make_blobs(n_samples = 500, n_features = 2, centers = 3, shuffle = True, random_state = 5)
```

```python
# plot
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
# Menambahkan argumen standar deviasi
X2, _ = make_blobs(n_samples = 500, n_features = 2, centers = 3, shuffle = True, cluster_std = 2, random_state = 5)
```

```python
# plot
plt.scatter(X2[:,0], X2[:,1])
plt.show()
```

## 3. Create a Model (KMeans)

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 3, init = 'random', max_iter = 500)
```

```python
y_km = km.fit_predict(X2)
```

```python
# Mencari posisi centroid
km.cluster_centers_ 
```

## 4. Visualisasi

```python
# Visualisasi Kluster
plt.scatter(X2[y_km==0,0], X2[y_km==0,1], label = 'Cluster 1')
plt.scatter(X2[y_km==1,0], X2[y_km==1,1], label = 'Cluster 2')
plt.scatter(X2[y_km==2,0], X2[y_km==2,1], label = 'Cluster 1')

# Visualisasi Centroid
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], c = 'red', label = 'Centroid')
plt.legend()
plt.show()
```

## 5. Elbow Method

Untuk mengetahui jumlah kluster terbaik yang direkomendasikan oleh komputer

```python
distortions = []
for i in range(1,11):
    km = KMeans(n_clusters = i, init = 'random', max_iter = 500)
    km.fit_predict(X2)
    distortions.append(km.inertia_)
```

```python
# Plot Elbow
plt.plot(range(1,11), distortions)
plt.xlabel('number of cluster')
plt.ylabel('distortions')

plt.show()
```
